```python
import os
import torch
import cv2
import shutil
from google.colab import drive


drive.mount('/content/drive')


!pip install torch==1.8.1 torchvision==0.9.1
!pip install opencv-python
!git clone https://github.com/ultralytics/yolov5
%cd yolov5
!pip install -r requirements.txt


model = torch.hub.load('ultralytics/yolov5', 'yolov5s')


def is_overlapping(box1, box2):
    """Verifica se box1 e box2 estão sobrepostos."""
    return (box1[0] < box2[2] and box1[2] > box2[0] and
            box1[1] < box2[3] and box1[3] > box2[1])


def draw_boxes(img, boxes, labels, colors):
    for i, box in enumerate(boxes):
        xmin, ymin, xmax, ymax = [int(x) for x in box]
        label = labels[i]
        color = colors[i]
        cv2.rectangle(img, (xmin, ymin), (xmax, ymax), color, 2)
        cv2.putText(img, label, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    return img


input_folder = '/content/drive/MyDrive/Sample_Yolo/dataSetTestYolo'
output_folder_with_bike = '/content/samples/with_bike'
output_folder_without_bike = '/content/samples/without_bike'


os.makedirs(output_folder_with_bike, exist_ok=True)
os.makedirs(output_folder_without_bike, exist_ok=True)


for image_name in os.listdir(input_folder):
    if image_name.lower().endswith(('.jpeg', '.jpg', '.webp')):
        image_path = os.path.join(input_folder, image_name)

        img = cv2.imread(image_path)

        results = model(img)

        detections = results.pandas().xyxy[0]

        people = detections[detections['name'] == 'person']
        bikes = detections[detections['name'] == 'bicycle']

        has_person_on_bike = False
        boxes = []
        labels = []
        colors = []
        for _, person in people.iterrows():
            person_box = (person['xmin'], person['ymin'], person['xmax'], person['ymax'])
            boxes.append(person_box)
            labels.append('person')
            colors.append((0, 255, 0))

            for _, bike in bikes.iterrows():
                bike_box = (bike['xmin'], bike['ymin'], bike['xmax'], bike['ymax'])
                if is_overlapping(person_box, bike_box):
                    has_person_on_bike = True
                boxes.append(bike_box)
                labels.append('bicycle')
                colors.append((255, 0, 0))

        img_with_boxes = draw_boxes(img, boxes, labels, colors)

        if has_person_on_bike:
            output_path = os.path.join(output_folder_with_bike, image_name)
        else:
            output_path = os.path.join(output_folder_without_bike, image_name)

        cv2.imwrite(output_path, img_with_boxes)
        print(f"Imagem {image_name} processada e salva em {output_path}")
```
